```python
import scipy as sp
from scipy import stats
import numpy as np
import pylab as pl
%pylab inline
```

```python
np.random.seed(125)
means = []
for ss in range(0,7):
  tmp = sp.stats.norm.rvs(size=10**ss)
  means.append(tmp.mean())
  print("sample mean ", means[-1])
  pl.figure()
  pl.hist(tmp)
  pl.show()
```

```python
pl.plot(means)
pl.axhline(0, color='k')
pl.text(0, 0.02, "population mean", ha="left", va="bottom")
pl.xlabel("sample size (log scale)")
pl.ylabel("sample mean");
```
