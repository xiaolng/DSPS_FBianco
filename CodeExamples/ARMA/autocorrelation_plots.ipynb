# autocorrelation tests

```python
import pandas as pd
import numpy as np
import pylab as pl
%pylab inline
```

## autocorrelation of a simple sine wave

```python
x = np.arange(100)
y = np.sin(x*0.1*np.pi)
```

```python
pl.plot(x,y, '.')
pl.plot(x,y, '-')
```

```python
matplotlib.pyplot.acorr(y, maxlags=90)
```

## autocorrelation of a double sine wave

```python
y = np.sin(x*0.1*np.pi + x*np.pi)
```

```python
pl.plot(x,y, '.')
pl.plot(x,y, '-')
```

```python
matplotlib.pyplot.acorr(y)
```

## using statsmodels: autocorrelation and partial autocorrelation

```python
import statsmodels.api as sm

from statsmodels.graphics.api import qqplot
```

```python
fig = sm.graphics.tsa.plot_acf(y, lags=80)
fig = sm.graphics.tsa.plot_pacf(y, lags=80)
```
